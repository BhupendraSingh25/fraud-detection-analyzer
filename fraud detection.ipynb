```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('loan.csv.csv')
df.head()
```

```python
df.loan_status.value_counts().plot(kind='bar')
plt.show()
```

```python
for i in df.columns:
    print(f'{i} <---->{df[i].isna().sum()/len(df)*100}')
```

```python
A = df.agg('mean')
df.fillna(value=A,axis=0,inplace=True)
df.isna().sum()
```

```python
df.dropna(axis=1,inplace=True)
df.isna().sum()
```

```python
df.head()
```

```python
num_col = df.select_dtypes(include=np.number).columns
cat_col = df.select_dtypes(exclude=np.number).columns
```

```python
for i in cat_col:
    print(df[i].value_counts())
    print('--'*50)
```

```python
for i in ['term','purpose','loan_status','grade','home_ownership','verification_status']:
    plt.pie(df[i].value_counts().values,labels=df[i].value_counts().index,autopct='%.2f%%')
    plt.show()
```

```python
for i in df.columns:
    df[i] = df[i].astype('category').cat.codes
df.head()
```

```python
df.drop(columns=['id','member_id'],inplace=True)
```

```python
plt.figure(figsize=(19,8))
corr = df.corr().abs()
corr['loan_status'].sort_values(ascending=False).plot(kind='bar')
plt.title('Range of Correlation With Drugs')
plt.show()
```
